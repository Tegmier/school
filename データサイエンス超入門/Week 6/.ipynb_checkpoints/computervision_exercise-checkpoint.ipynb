```python
# ライブラリをimportする
import cv2
from PIL.Image import fromarray
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
#画像表示用関数
def imshow(img):
    ret, encoded = cv2.imencode(".jpg", img)
    display(Image(encoded))
```

```python
# 画像読み込み
im = cv2.imread('bread.jpg')
# 画像を表示
imshow(im)
```

```python
#グレイスケール化
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
imshow(im_gray)
```

```python
#ラプラシアンフィルタ
im_edge = cv2.Laplacian(im, cv2.CV_32F)
imshow(im_edge)
```

```python
anpan = im[310:500, 500:700]
imshow(anpan)
```

```python
color = ('b','g','r')
for channel,col in enumerate(color):
    histr = cv2.calcHist([anpan],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('anpan')
plt.show()
```

```python
choco = im[100:310, 300:500]
imshow(choco)
```

```python
color = ('b','g','r')
for channel,col in enumerate(color):
    histr = cv2.calcHist([choco],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('choco')
plt.show()
```

```python
# 画像読み込み
shibuya = cv2.imread('shibuya.jpg')
color = ('b','g','r')
for channel,col in enumerate(color):
    histr = cv2.calcHist([shibuya],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('shibuya')
plt.show()
```

```python
#画像読み込み
im = cv2.imread('bread.jpg')
#sft計算
sift = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(im, None)
im_sift = cv2.drawKeypoints(im, keypoints, None, flags=4)
#描画
imshow(im_sift)
```
