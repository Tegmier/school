```python
import glob
import pandas as pd
```

```python
# Get list of files that ends with .csv
file_list = glob.glob("data/All_20053_20232/*.csv")
# sort files
file_list.sort()
```

```python
file_list
```

```python
%%time
# Loop through files
list_df = []
for i in range(len(file_list)):
    df_tmp = pd.read_csv(file_list[i],encoding="cp932")
    list_df.append(df_tmp)
```

```python
list_df[0].head(10)
```

```python
print(len(list_df))
```

```python
df_all = pd.concat(list_df)
```

```python
length = 0
for i in range(len(file_list)):
    length += len(list_df[i])
```

```python
len(df_all) == length
```

Sampe and Save

```python
df_sample = df_all.sample(100000, random_state = 123)
```

```python
len(df_sample)
```

```python
df_sample.to_csv('data/SuperD_Class2_tochi_sample_xdm.csv', index = False)
```
