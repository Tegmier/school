Longitude and Latitude

```python
df_code = pd.read_csv("data/SuperD_Class2_shicyoukuson_code_utf8.csv")
```

```python
df_code.head()
```

```python
code2lng, code2lat = ({} for _ in range(2))
```

```python
count = 0
for i in range(len(df["市区町村コード"])):
    if df.iloc[i]["市区町村コード"] not in df_code["コード"]:
        count +=1
count
```

```python
df.drop_duplicates(subset = "市区町村コード")["市区町村コード"]
```

```python
(df["市区町村コード"]/10000 < 1).value_counts()
```

```python
for i in range(len(df_code)):
    code = str(df_code["コード"].iloc[i])
    if (len(code) == 5):
        code = code[0:4]
    else:
        code = code[0:5]
    code2lng.update({int(code) : df_code["経度"].iloc[i]})
    code2lat.update({int(code) : df_code["緯度"].iloc[i]})
```

```python
df.isnull().sum() / len(df)
```

```python
#load the coordinate data to the data frame
df["longitude"] = df["市区町村コード"].map(code2lng)
df["latitude"] = df["市区町村コード"].map(code2lat)
df
```

```python
#To plot the map and make sure it looks lie the map of Japan
df.plot(x = "longitude", y = "latitude",kind = 'scatter', label = 'Housing Location', marker = '+')
plt.title("Scatterplot of all transations")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.show()
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn')
import numpy as np
inertia_values = {}
np_array = df[["longitude", "latitude"]].values
for k in range(2,30):
    k_means = KMeans(n_clusters = k, random_state = 42)
    k_means.fit(np_array)
    inertia_values.update({k : k_means.inertia_})
```

```python
min(inertia_values.keys())
```

```python
k_means = KMeans(n_clusters = 10, random_state = 42)
k_means.fit(np_array)
centroids = k_means.cluster_centers_
labels = k_means.labels_
```

```python
plt.figure(figsize=(10,10),dpi=200)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.scatter(np_array[:, 0], np_array[:, 1],marker = 1,s=8)
plt.scatter(centroids[:,0],centroids[:,1],color="red",marker=6,s=48)
plt.show()
```
